```python
import pandas as pd
import numpy as np

df = pd.read_excel("World_development_mesurement (1) (1).xlsx")
df.head(10)
```

```python
df1 = df.drop(['Business Tax Rate','Ease of Business','Hours to do Tax','Number of Records','Country'], axis = 1, errors='ignore')
```

```python
df1.sample(5)
```

```python
df1.isnull().sum()
```

```python
for col in df1.columns:
    if df1[col].dtype == 'object' and col != 'Country': 
        df1[col] = df1[col].str.replace('$', '', regex=False) #remove '$' From values
        df1[col] = df1[col].str.replace(',', '', regex=False) #remove ',' from values
        df1[col] = pd.to_numeric(df1[col], errors='coerce')   #convert column into numeric dtype
```

```python
df1.sample(10)
```

```python
df1.dtypes
```

```python
df1.isnull().sum()
```

```python
df1 = df1.fillna(df1.median(numeric_only=True)) # Fill Null values With Median 
```

```python
df1.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = df1.select_dtypes(include=[np.number]).columns.tolist() #Find Numerical Columns 
plt.figure(figsize=(15, 15))
for i, col in enumerate(num_cols, 1):
    plt.subplot(5, 4, i)
    sns.boxplot(x=df[col], color="skyblue")  #Plot Boxplot For Outlier Detection 
    plt.title(f"Boxplot: {col}")
plt.tight_layout()
plt.show()
```

```python
df2 = df1.copy()

for col in num_cols :
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1 

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df2[(df2[col] < lower_bound) | (df2[col] > upper_bound)]
    print(f'{len(outliers)} Outliers in {col}')
```

```python
for col in num_cols :
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1 

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df2[col] = df2[col].apply(lambda x: max(min(x, upper_bound), lower_bound)) # capping For handel Outliers 
```

```python
plt.figure(figsize = (20,5))
sns.boxplot(df2)
plt.show()
```
