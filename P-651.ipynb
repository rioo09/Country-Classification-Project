```python
df1 = df.drop(['Business Tax Rate','Ease of Business','Hours to do Tax','Number of Records','Country'], axis = 1, errors='ignore') #Dropping unnecessary columns from the dataset
```

```python
df1.sample(5)
```

```python
for col in df1.columns:
    if col != 'Country':
        df1[col] = pd.to_numeric(                   #Convert column to numeric type
            df1[col].astype(str)                    #Convert entire column to string type
            .str.replace('$', '', regex=False)      #Remove '$' from values
            .str.replace(',', '', regex=False),     #Remove ',' from values
            errors='coerce'                         #Convert to numeric, setting non-convertible values to NaN
        )
```

```python
df1 = df1.fillna(df1.median(numeric_only=True)) # Fill Null values With Median 
```

```python
df1.isnull().sum() #Checking for null values after filling them with median
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = df1.select_dtypes(include=[np.number]).columns.tolist() #Find Numerical Columns 

plt.figure(figsize=(15, 15))
for i, col in enumerate(num_cols, 1):
    plt.subplot(5, 4, i)
    sns.boxplot(x=df[col], color="skyblue")  #Plot Boxplot For Outlier Detection 
    plt.title(f"Boxplot: {col}")
plt.tight_layout()
plt.show()
```

```python
df2 = df1.copy()

for col in num_cols :
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1 

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df2[(df2[col] < lower_bound) | (df2[col] > upper_bound)] #Identify Outliers Based on IQR Method
    print(f'{len(outliers)} Outliers in {col}')
```

```python
for col in num_cols :
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1 

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df2[col] = df2[col].apply(lambda x: max(min(x, upper_bound), lower_bound)) # capping For handel Outliers 
```

```python
plt.figure(figsize = (20,5))
sns.boxplot(df2) #Boxplot after handel outliers with capping method
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sd = StandardScaler()
df_scaled = sd.fit_transform(df2) #Scaling the dataset using StandardScaler
df_scaled = pd.DataFrame(df_scaled, columns=df2.columns) #Converting Into DataFrame After Scaling
df_scaled.head()
```

```python
plt.figure(figsize = (20,10))
sns.boxplot(df_scaled) 
```

```python
from sklearn.decomposition import PCA #import PCA
pca = PCA(n_components=20)
pcomp = pca.fit_transform(df_scaled) #fit_transform df_scaled using PCA
```

```python
col = []
for i in range(1,21):
    col.append('PC'+str(i))  #Create Column Names For PCA Components
df_pca = pd.DataFrame(pcomp, columns=col) #Convert PCA Components Into DataFrame With Column Names
df_pca.head()
```

```python
exl_var = pca.explained_variance_ratio_  #Calculate Explained Variance Ratio For Each PCA Component
np.round(exl_var, 3)
```

```python
np.cumsum(exl_var).round(3)  #cululative sum of explained variance ratio
```

```python
final_df = df_pca.iloc[:,:11] #Selecting the first 11 PCA components which explain around 95% of the variance in the data
final_df.head()
```

```python
plt.figure(figsize = (6,6))
sns.heatmap(final_df.corr()>0.7, annot = True, cmap = 'coolwarm') #Plot Heatmap To Identify Correlation Between PCA Components
plt.show()
```

```python
from sklearn.manifold import TSNE                                       #importting TSNE
tsne = TSNE(n_components= 2, perplexity=40)                             
tsne_comp = tsne.fit_transform(final_df)                                #fit_transform final_df using TSNE
tsne_df = pd.DataFrame(tsne_comp, columns=['TSNE1', 'TSNE2'])           #Convert TSNE Components Into DataFrame
tsne_df.head()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data= tsne_df, x='TSNE1', y='TSNE2' ) #Plotting t-SNE Visualization
plt.title("t-SNE Visualization")
plt.show()
```

### Model Building

```python
from sklearn.cluster import KMeans #importing KMeans

inertia = []
for k in range(1,11):
    kmeans = KMeans(n_clusters = k, random_state = 42)
    kmeans.fit(final_df)
    inertia.append(kmeans.inertia_) #Calculate Inertia For Each Value Of K

plt.figure(figsize = (8,5))
plt.plot(range(1,11), inertia, marker = "o")  #Plotting Elbow Method 
plt.title("Elbow Method For Otimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")   
plt.show()
```

```python
km = KMeans(n_clusters = 4, random_state = 42)
km.fit(final_df) #fit final_df using KMeans
pred = km.predict(final_df) #Predict Cluster Labels
```

```python
from sklearn.metrics import silhouette_score 
silhouette_avg = silhouette_score(final_df, km.labels_) #Calculate Silhouette Score For KMeans Clustering
print(f'Silhouette Score: {silhouette_avg}')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = final_df['PC1'], y = final_df['PC2'], hue=pred, palette='Set1') #Plotting Clusters Based On First Two PCA Components
plt.title("KMeans Clusters Based on PCA Components")
```
