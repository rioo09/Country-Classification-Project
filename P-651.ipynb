```python
for col in df1.columns:
    if col != 'Country':
        df1[col] = pd.to_numeric(                   #Convert column to numeric type
            df1[col].astype(str)                    #Convert entire column to string type
            .str.replace('$', '', regex=False)      #Remove '$' from values
            .str.replace(',', '', regex=False),     #Remove ',' from values
            errors='coerce'                         #Convert to numeric, setting non-convertible values to NaN
        )
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_cols = df1.select_dtypes(include=[np.number]).columns.tolist() #Find Numerical Columns 

plt.figure(figsize=(15, 15))
for i, col in enumerate(num_cols, 1):
    plt.subplot(5, 4, i)
    sns.boxplot(x=df[col], color="skyblue")  #Plot Boxplot For Outlier Detection 
    plt.title(f"Boxplot: {col}")
plt.tight_layout()
plt.show()
```

```python
df2 = df1.copy()

for col in num_cols :
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1 

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df2[(df2[col] < lower_bound) | (df2[col] > upper_bound)] #Identify Outliers Based on IQR Method
    print(f'{len(outliers)} Outliers in {col}')
```

```python
for col in num_cols :
    Q1 = df2[col].quantile(0.25)
    Q3 = df2[col].quantile(0.75)
    IQR = Q3 - Q1 

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df2[col] = df2[col].apply(lambda x: max(min(x, upper_bound), lower_bound)) # capping For handel Outliers 
```

```python
plt.figure(figsize = (20,5))
sns.boxplot(df2) #Boxplot after handel outliers with capping method
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
sd = StandardScaler()
df_scaled = sd.fit_transform(df2) #Scaling the dataset using StandardScaler
df_scaled = pd.DataFrame(df_scaled, columns=df2.columns) #Converting Into DataFrame After Scaling
df_scaled.head()
```

```python
from sklearn.decomposition import PCA #import PCA
pca = PCA(n_components=20)
pcomp = pca.fit_transform(df_scaled) #fit_transform df_scaled using PCA
```

```python
col = []
for i in range(1,21):
    col.append('PC'+str(i))  #Create Column Names For PCA Components
    
df_pca = pd.DataFrame(pcomp, columns=col) #Convert PCA Components Into DataFrame With Column Names
df_pca.head()
```

```python
exl_var = pca.explained_variance_ratio_  #Calculate Explained Variance Ratio For Each PCA Component
np.round(exl_var, 3)
```

```python
np.cumsum(exl_var).round(3)  #cululative sum of explained variance ratio
```

```python
final_df = df_pca.iloc[:,:11] #Selecting the first 11 PCA components which explain around 95% of the variance in the data
final_df.head()
```

```python
plt.figure(figsize = (6,6))
sns.heatmap(final_df.corr()>0.7, annot = True, cmap = 'coolwarm') #Plot Heatmap To Identify Correlation Between PCA Components
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data= tsne_df, x='TSNE1', y='TSNE2' ) #Plotting t-SNE Visualization
plt.title("t-SNE Visualization")
plt.show()
```

# Model Building

```python
from sklearn.cluster import KMeans #importing KMeans

inertia = []
for k in range(1,11):
    kmeans = KMeans(n_clusters = k, random_state = 42)
    kmeans.fit(final_df)
    inertia.append(kmeans.inertia_) #Calculate Inertia For Each Value Of K

plt.figure(figsize = (8,5))
plt.plot(range(1,11), inertia, marker = "o")  #Plotting Elbow Method 
plt.title("Elbow Method For Otimal K")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")   
plt.show()
```

# KMeans

```python
km = KMeans(n_clusters = 3, random_state = 42)
km.fit(final_df) #fit final_df using KMeans
pred_km = km.predict(final_df) #Predict Cluster Labels
```

```python
from sklearn.metrics import silhouette_score 
silhouette_avg = silhouette_score(final_df, km.labels_) #Calculate Silhouette Score For KMeans Clustering
print(f'Silhouette Score: {silhouette_avg}')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = final_df['PC1'], y = final_df['PC2'], hue=pred_km, palette='Set1') #Plotting Clusters Based On First Two PCA Components
plt.title("KMeans Clusters Based on PCA Components")
```

# DBSCAN

```python
from sklearn.cluster import DBSCAN

eps_values = [0.3,1.5, 2, 2.5, 3, 3.5, 4, 4.5]
min_samples_values = [3, 5, 10, 15]

for eps in eps_values:
    for min_samples in min_samples_values:
        db = DBSCAN(eps=eps, min_samples=min_samples)
        labels = db.fit_predict(final_df)
        n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
        sil_score = -1
        if n_clusters > 1:
            sil_score = silhouette_score(final_df, labels)
        print(f"EPS: {eps}, MinPts: {min_samples}, Clusters: {n_clusters}, Silhouette Score: {sil_score:.3f}")
```

```python
db = DBSCAN(eps=2.5, min_samples=3)
pred_db= db.fit_predict(final_df) #fit_predict final_df using DBSCAN
```

```python
silhouette_avg = silhouette_score(final_df, pred_db) #Calculate Silhouette Score For DBScan Clustering
print(f'Silhouette Score: {silhouette_avg}')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = final_df['PC1'], y = final_df['PC2'], hue=pred_db, palette='Set1') #Plotting Clusters Based On First Two PCA Components
plt.title("DBSCAN Clusters Based on PCA Components")
```

# Hierarchical clustering / Agglomerative clustering

```python
from sklearn.cluster import AgglomerativeClustering

linkages = ['ward','complete','average','single']

for linkage in linkages:
    for k in range(2,5):   # testing clusters from 2 to 4
        
        # ward only works with euclidean distance
        if linkage == 'ward':
            model = AgglomerativeClustering(n_clusters=k, linkage=linkage)
        else:
            model = AgglomerativeClustering(n_clusters=k, linkage=linkage, metric='euclidean')
        
        labels = model.fit_predict(final_df)
        score = silhouette_score(final_df, labels)
        
        print(f"Linkage: {linkage}, Clusters: {k}, Silhouette Score: {score:.3f}")
```

```python
agg = AgglomerativeClustering(n_clusters=2, linkage='average') #Create AgglomerativeClustering Model With Optimal Parameters
pred_agg = agg.fit_predict(final_df)
```

```python
silhouette_avg = silhouette_score(final_df, pred_agg) #Calculate Silhouette Score For Agglomerative Clustering
print(f'Silhouette Score: {silhouette_avg}')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = final_df['PC1'], y = final_df['PC2'], hue=pred_agg, palette='Set1') #Plotting Clusters Based On First Two PCA Components
plt.title("Agglomerative Clusters Based on PCA Components")
```

# Gaussian Mixture Model (GMM)

```python
from sklearn.mixture import GaussianMixture

cov_types = ['full','tied','diag','spherical']

for cov in cov_types:
    for k in range(2,5):   # testing clusters from 2 to 4
        
        gmm = GaussianMixture(
            n_components=k,
            covariance_type=cov,
            random_state=42
        )
        
        labels = gmm.fit_predict(final_df)
        score = silhouette_score(final_df, labels)
        
        print(f"Covariance: {cov}, Clusters: {k}, Score: {score:.3f}")
```

```python
gmm = GaussianMixture(n_components=3, covariance_type='spherical', random_state=42)
pred_gmm = gmm.fit_predict(final_df)
```

```python
silhouette_avg = silhouette_score(final_df, pred_gmm) #Calculate Silhouette Score For Gaussian Mixture Model Clustering
print(f'Silhouette Score: {silhouette_avg}')
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x = final_df['PC1'], y = final_df['PC2'], hue=pred_gmm, palette='Set1') #Plotting Clusters Based On First Two PCA Components
plt.title("GMM Clusters Based on PCA Components")
```

# Model Comparison

```python
models = ['KMeans', 'DBSCAN', 'Agglomerative', 'GMM']

Silhouettes = [
    silhouette_score(final_df, pred_km),
    silhouette_score(final_df, pred_db),
    silhouette_score(final_df, pred_agg),
    silhouette_score(final_df, pred_gmm)
]  #Silhouette Scores For Each Clustering Model

plt.figure(figsize=(10,6))
sns.set_style("whitegrid")
ax = sns.barplot(x=models, y=Silhouettes, palette="viridis")
plt.title("Comparison of Clustering Algorithms using Silhouette Score", fontsize=14)
plt.xlabel("Clustering Algorithms", fontsize=12)
plt.ylabel("Silhouette Score", fontsize=12)

for i, v in enumerate(Silhouettes):
    ax.text(i, v + 0.01, f"{v:.2f}", ha='center', fontsize=11) # Add score labels on bars
plt.ylim(0, max(Silhouettes) + 0.1)
plt.tight_layout()
plt.show()
```
